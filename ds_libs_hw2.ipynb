## Задание 2. График косинусов

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)

plt.plot(t, f, color='#00FF00')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f(t)')
plt.axis([0.5, 9.5, -2.5, 2.5])
```

## *Задание 3. График четырех функций

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

fig, axes = plt.subplots(figsize=(8, 6), ncols=2, nrows=2)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

axes[0, 0].plot(x, y1)
axes[0, 0].set_title('График y1')
axes[0, 0].set_xlim([-5, 5])

axes[0, 1].plot(x, y2)
axes[0, 1].set_title('График y2')

axes[1, 0].plot(x, y3)
axes[1, 0].set_title('График y3')

axes[1, 1].plot(x, y4)
axes[1, 1].set_title('График y4')
```
