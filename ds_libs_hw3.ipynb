# Задания

## Задание 1. Обучение модели линейной регрессии

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
lr_r2 = r2_score(y_test, y_pred)

plt.figure(figsize=(10, 10))
ax = sns.scatterplot(x=y_pred[:, 0], y=y_test.values[:, 0])
sns.lineplot(x=[0, 50], y=[0, 50], linestyle='--')
ax.set(xlabel='y_pred', ylabel='y_test')
ax.text(0, 50, f'R2 = {lr_r2:.3f}')
plt.title('Linear regression')
```

Примечательно, что по одному дому предсказанная цена получилась отрицательной

## Задание 2. Обучение модели случайного леса

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
model.fit(x_train, y_train.values[:, 0])
```

```python
y_pred = model.predict(x_test)
forest_r2 = r2_score(y_test, y_pred)

plt.figure(figsize=(10, 10))
ax = sns.scatterplot(x=y_pred, y=y_test.values[:, 0])
sns.lineplot(x=[0, 50], y=[0, 50])
ax.set(xlabel='y_pred', ylabel='y_test')
ax.text(0, 50, f'R2 = {forest_r2:.3f}')
plt.title('RandomForestRegressor')
```

Случайный лес получился гораздо точнее линейной регрессии

```python
print(f'Linear regression R2: {lr_r2:.3f}')
print(f'Random forest R2: {forest_r2:.3f}')
```
